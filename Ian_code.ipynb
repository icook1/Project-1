```python
final_data_merge["Happiness Score_x_x"] = final_data_merge["Happiness Score_x_x"].astype(str)
final_data_merge["Happiness Score_y_x"] = final_data_merge["Happiness Score_y_x"].astype(str)
final_data_merge["Happiness Score_x_y"] = final_data_merge["Happiness Score_x_y"].astype(str)
final_data_merge["Happiness Score_y_y"] = final_data_merge["Happiness Score_y_y"].astype(str)
final_data_merge["Happiness Score"] = final_data_merge["Happiness Score"].astype(str)
```

```python
final_data_merge["Happiness Score_x_x"] = final_data_merge["Happiness Score_x_x"].str.replace(',', '')
final_data_merge["Happiness Score_y_x"] = final_data_merge["Happiness Score_y_x"].str.replace(',', '')
final_data_merge["Happiness Score_x_y"] = final_data_merge["Happiness Score_x_y"].str.replace(',', '')
final_data_merge["Happiness Score_y_y"] = final_data_merge["Happiness Score_y_y"].str.replace(',', '')
final_data_merge["Happiness Score"] = final_data_merge["Happiness Score"].str.replace(',', '')
```

```python
df=final_data_merge.loc[:,["Country","Happiness Score_x_x", "Happiness Score_y_x", "Happiness Score_x_y", "Happiness Score_y_y", "Happiness Score"]]
df= df.rename(columns={'Happiness Score_x_x': 'Happiness Score 2018', 'Happiness Score_y_x': 'Happiness Score 2019', 'Happiness Score_x_y': 'Happiness Score 2020', 'Happiness Score_y_y': 'Happiness Score 2021', 'Happiness Score': 'Happiness Score 2022' })
df
```

```python
df['Happiness Score 2018'] = pd.to_numeric(df['Happiness Score 2018'], errors='coerce').astype(float)
df['Happiness Score 2019'] = pd.to_numeric(df['Happiness Score 2019'], errors='coerce').astype(float)
df.dropna(subset=["Happiness Score 2018", "Happiness Score 2019"], inplace=True)
```

```python
correlation = st.pearsonr(df['Happiness Score 2018'], df['Happiness Score 2019'])
print(f"Correlation of a Countries Happiness score between 2018 and 2019 {round(correlation[0],2)}")
```

```python
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(df["Happiness Score 2018"], df["Happiness Score 2019"])
regress = df["Happiness Score 2018"]* slope + intercept
line = f"y = {round(slope, 2)} x + {round(intercept, 2)}"

plt.scatter(df["Happiness Score 2018"], df["Happiness Score 2019"], color="orange")
plt.plot(df["Happiness Score 2018"], regress, color = "blue")
plt.annotate(line, (22,37), fontsize=12)
plt.xlabel("Happiness Score 2018")
plt.ylabel("Happiness Score 2019")
plt.title("Happiness Scores 2018 v. 2019")
print(f"R-squared is: {rvalue**2}")
plt.show()
```

```python
df['Happiness Score 2019'] = pd.to_numeric(df['Happiness Score 2019'], errors='coerce').astype(float)
df['Happiness Score 2020'] = pd.to_numeric(df['Happiness Score 2020'], errors='coerce').astype(float)
df.dropna(subset=["Happiness Score 2019", "Happiness Score 2020"], inplace=True)
```

```python
start_column = 1
end_column = 6

df.set_index(df.columns[0], inplace=True)

selected_columns = df.iloc[:, start_column:end_column+1]

max_values = selected_columns.max(axis=1)
max_column_indices = selected_columns.idxmax(axis=1)

for idx, (max_value, max_column_index) in zip(max_values.index, zip(max_values, max_column_indices)):
    print(f"Row {idx}: Maximum Value = {max_value}, Column Index = {max_column_index}")
```

```python
happiness_df = pd.DataFrame({'Country': max_values.index, 'Maximum Value': max_values, 'Happiness Score Year': max_column_indices})
print(happiness_df)
```

```python


limit_df = happiness_df.head(25)


limit_df.plot(x='Country', y='Maximum Value', kind='bar', title='Maximum Happiness Score by Top 25 Countries')



plt.figure(figsize=(12, 6))
plt.bar(limit_df['Country'], limit_df['Maximum Value'])
plt.xlabel('Country')
plt.ylabel('Happiness Score Max')
plt.title('Maximum Happiness Score by Top 25 Countries')
plt.xticks(rotation=90)

plt.show()
```

```python
happiness_year_counts = happiness_df['Happiness Score Year'].value_counts()

# Plotting the count of Happiness Score Year using a bar graph
plt.figure(figsize=(10, 6))
happiness_year_counts.plot(kind='bar', color='skyblue')
plt.title('Count of Max Happiness Score Year')
plt.xlabel('Happiness Score Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

```python
limit_df = happiness_df.tail(25)


limit_df.plot(x='Country', y='Maximum Value', kind='bar', title='Minimum Happiness Score by Bottom 25 Countries')



plt.figure(figsize=(12, 6))
plt.bar(limit_df['Country'], limit_df['Maximum Value'])
plt.xlabel('Country')
plt.ylabel('Happiness Score Max')
plt.title('Minimum Happiness Score by Bottom 25 Countries')
plt.xticks(rotation=90)

plt.show()
```
